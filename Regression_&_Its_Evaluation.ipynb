Q.2 What are the key assumptions of Simple Linear Regression...?
Ans. 1. Linearity
- The relationship between the independent variable (X) and the dependent variable (Y) must be linear.
- In other words, changes in X result in proportional changes in Y.
2. Independence
- The residuals (errors) should be independent.
- This means that the outcome of one observation should not influence another.
- Violations often occur with time-series data (e.g., autocorrelation).
3. Homoscedasticity
- The variance of residuals should remain constant across all levels of X.
- In simpler terms: your prediction errors should be evenly spread.
4. Normality of Residuals
- The residuals should be approximately normally distributed.
- This is especially important for hypothesis testing (e.g., confidence intervals, p-values).
5. No or Little Multicollinearity (more relevant for multiple regression)
- In SLR, there's just one predictor, so multicollinearity isn’t a concern.
- But it becomes crucial when you're dealing with multiple predictors.

Q.3  What is heteroscedasticity, and why is it important to address in regression models..?
Ans. In a linear regression model, one of the key assumptions is that the residuals (errors) have constant variance — this is known as homoscedasticity. When this assumption is violated (i.e., variance of the errors increases or decreases with the value of an independent variable), it results in heteroscedasticity.

You can visualize it as a funnel-shaped pattern in a residual vs. fitted values plot:

Narrow at one end (small variance),

Wide at the other (large variance).

⚠️ Why Heteroscedasticity Is a Problem:
Invalid Standard Errors:

Leads to biased or inconsistent estimates of standard errors, which affects hypothesis testing (t-tests, F-tests).

May result in incorrect p-values, leading you to accept or reject hypotheses incorrectly.

Inefficient Estimates:

The OLS estimators remain unbiased but are no longer BLUE (Best Linear Unbiased Estimators).

There's loss of efficiency, meaning better (more precise) estimates could exist.

Misleading Confidence Intervals:

Confidence intervals become unreliable, possibly too wide or too narrow.

✅ How to Detect Heteroscedasticity:
Graphical Method:

Residuals vs. Fitted values plot (look for a funnel shape).

Statistical Tests:

Breusch–Pagan test

White test

Goldfeld–Quandt test

🛠 How to Address Heteroscedasticity:
Transform the Dependent Variable:

Use a logarithmic, square root, or Box-Cox transformation to stabilize variance.

Use Weighted Least Squares (WLS):

Assigns weights to data points based on variance to correct for heteroscedasticity.

Use Robust Standard Errors:

Adjust standard errors to remain valid even under heteroscedasticity (e.g., White’s robust standard errors).

```python
'''Q.6  Implement a Python program to fit a Simple Linear Regression model to
the following sample data:
● X = [1, 2, 3, 4, 5]
● Y = [2.1, 4.3, 6.1, 7.9, 10.2]
Plot the regression line over the data points.'''
"Ans."
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = [1, 2, 3, 4, 5]
Y = [2.1, 4.3, 6.1, 7.9, 10.2]

# Convert to numpy arrays and reshape X for sklearn
X = np.array(X).reshape(-1, 1)
Y = np.array(Y)

# Create and fit the model
model = LinearRegression()
model.fit(X, Y)

# Predict Y values
Y_pred = model.predict(X)

# Plot original data points
plt.scatter(X, Y, color='blue', label='Actual Data')

# Plot regression line
plt.plot(X, Y_pred, color='red', label='Regression Line')

# Add labels and legend
plt.title('Simple Linear Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()

# Print model parameters
print(f"Intercept (β₀): {model.intercept_}")
print(f"Slope (β₁): {model.coef_[0]}")
```

```python
''' Q.7 Fit a Multiple Linear Regression model on this sample data:
● Area = [1200, 1500, 1800, 2000]
● Rooms = [2, 3, 3, 4]
● Price = [250000, 300000, 320000, 370000]
Check for multicollinearity using VIF and report the results. '''
"Ans.7"
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm

# Sample data
data = {
    'Area': [1200, 1500, 1800, 2000],
    'Rooms': [2, 3, 3, 4],
    'Price': [250000, 300000, 320000, 370000]
}

df = pd.DataFrame(data)

# Define features and target
X = df[['Area', 'Rooms']]
y = df['Price']

# Fit the regression model
model = LinearRegression()
model.fit(X, y)

# Print model coefficients
print(f"Intercept (β₀): {model.intercept_}")
print(f"Coefficients (β): {model.coef_}")


```

```python
'''Q.8  Implement polynomial regression on the following data:
● X = [1, 2, 3, 4, 5]
3
● Y = [2.2, 4.8, 7.5, 11.2, 14.7]
Fit a 2nd-degree polynomial and plot the resulting curve. '''
"Ans.8"
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# Transform features to 2nd-degree polynomial
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Fit the polynomial regression model
model = LinearRegression()
model.fit(X_poly, Y)

# Predict values
X_fit = np.linspace(1, 5, 100).reshape(-1, 1)
X_fit_poly = poly.transform(X_fit)
Y_pred = model.predict(X_fit_poly)

# Plot original data points
plt.scatter(X, Y, color='blue', label='Actual Data')

# Plot polynomial regression curve
plt.plot(X_fit, Y_pred, color='red', label='2nd-degree Polynomial Fit')

# Add titles and labels
plt.title('Polynomial Regression (Degree 2)')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()

# Print model coefficients
print(f"Intercept (β₀): {model.intercept_}")
print(f"Coefficients (β₁, β₂): {model.coef_[1:]}")
```

```python
'''Q.9  Create a residuals plot for a regression model trained on this data:
● X = [10, 20, 30, 40, 50]
● Y = [15, 35, 40, 50, 65]
Assess heteroscedasticity by examining the spread of residuals. '''
"Ans.9"
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

# Fit linear regression model
model = LinearRegression()
model.fit(X, Y)

# Predict Y values
Y_pred = model.predict(X)

# Calculate residuals
residuals = Y - Y_pred

# Plot residuals vs. predicted values
plt.scatter(Y_pred, residuals, color='purple')
plt.axhline(y=0, color='black', linestyle='--')  # Horizontal line at 0
plt.title('Residuals Plot')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals (Actual - Predicted)')
plt.grid(True)
plt.show()

# Optional: print residuals
print("Predicted Values:", Y_pred)
print("Residuals:", residuals)
```

need to predict house prices using features like area, number of rooms, and location.
However, you detect heteroscedasticity and multicollinearity in your regression
model. Explain the steps you would take to address these issues and ensure a robust
model. '''
"Ans.10"
 1. Detect the Issues Clearly
Heteroscedasticity
Plot residuals vs. predicted values → look for patterns or changing spread.

Run formal tests: Breusch–Pagan, White test.

Multicollinearity
Calculate VIF (Variance Inflation Factor) for all features.

VIF > 5 → moderate concern

VIF > 10 → serious multicollinearity

🔧 2. Fixing Heteroscedasticity (Non-constant error variance)
🔄 a. Transform the Target Variable
Apply log, square root, or Box-Cox transformation on house prices:

Edit
y = np.log(price)  # for example
This stabilizes the variance and often improves linearity too.

⚖️ b. Use Weighted Least Squares (WLS)
Assign weights to each observation inversely proportional to their variance.

Helps give less weight to high-variance (outlier) observations.

🛡️ c. Use Robust Standard Errors
In statsmodels, use:

python
Copy
Edit
model = sm.OLS(y, X).fit(cov_type='HC3')
This adjusts standard errors to be reliable even under heteroscedasticity.

🔧 3. Fixing Multicollinearity (Highly correlated predictors)
🔍 a. Remove or Combine Correlated Features
Example: Area and Number of Rooms may be strongly correlated.

Consider removing one, or creating a new variable (e.g., area per room).

📉 b. Apply Dimensionality Reduction
Use PCA (Principal Component Analysis) to combine correlated features into uncorrelated components (though interpretability is reduced).

🧠 c. Regularization Techniques
Use Ridge Regression (L2) or Lasso Regression (L1):

python
Copy
Edit
from sklearn.linear_model import Ridge, Lasso
model = Ridge(alpha=1.0).fit(X, y)
These penalize large coefficients and reduce overfitting from multicollinearity.

✅ 4. Evaluate and Validate the Model
Use cross-validation to check performance on unseen data.

Monitor metrics like:

R² / Adjusted R²

RMSE / MAE

Check residual plots again after transformation or regularization.
